## Question 2: Filter Products by Category

This endpoint allows users to filter products based on their category.

The category name is provided as a path parameter in the API URL.
The API returns all products that belong to the specified category.

Example:
- /products/category/Electronics
- /products/category/Stationery

The filtering is case-insensitive to make the search more flexible.

```python
# ===============================
# Q2: Products by category
# ===============================

@app.get("/question2/{category_name}")

def question2(category_name: str):
    category_name = category_name.strip()  # extra spaces remove
    filtered = [p for p in products if p["category"].lower() == category_name.lower()]
    if not filtered:
        return {"message": f"No products found in category '{category_name}'"}
    return {"category": category_name, "total": len(filtered), "products": filtered}
```

## Question 3: Get In-Stock Products

This endpoint returns only those products that are currently available in stock.

The API filters the product list and selects only those items where the
`in_stock` value is True.

This helps users easily identify which products are available for purchase.

```python
# ===============================
# Q3: Only in-stock products
# ===============================

@app.get("/question3")

def question3():
    instock = [p for p in products if p["in_stock"]]
    return {"in_stock_count": len(instock), "in_stock_products": instock}
```

## Question 4: Store Summary

This endpoint provides a quick overview of the store inventory.

The summary includes:
- Total number of products
- Number of products currently in stock
- Number of products out of stock
- List of unique product categories

This information helps in understanding the current status of the store.

```python
# ===============================
# Q4: Store summary
# ===============================

@app.get("/question4")

def question4():
    total = len(products)
    in_stock = len([p for p in products if p["in_stock"]])
    out_stock = total - in_stock
    categories = list(set(p["category"] for p in products))
    return {
        "store_name": "My E-commerce Store",
        "total_products": total,
        "in_stock": in_stock,
        "out_of_stock": out_stock,
        "categories": categories
    }
```

## Question 5: Search Products

This endpoint allows users to search for products using a keyword.

The API checks whether the given keyword exists in the product name.
The search is case-insensitive, meaning it will match both uppercase
and lowercase characters.

Example searches:
- /products/search/mouse
- /products/search/book

```python
# ===============================
# Q5: Search products by keyword
# ===============================

@app.get("/question5/{keyword}")

def question5(keyword: str):
    keyword = keyword.strip()
    matches = [p for p in products if keyword.lower() in p["name"].lower()]
    if not matches:
        return {"message": f"No products matched keyword '{keyword}'"}
    return {"keyword": keyword, "total_matches": len(matches), "products": matches}
```

## Question 6 (Bonus): Best Deals

This endpoint identifies the best deals available in the store.

It returns:
- The cheapest product (best deal)
- The most expensive product (premium pick)

The `min()` and `max()` functions are used with a lambda expression
to determine the products based on their price.

```python
# ===============================
# Q6: Cheapest, Most Expensive & Average Price
# ===============================

@app.get("/question6")

def question6():
    cheapest = min(products, key=lambda p: p["price"])
    expensive = max(products, key=lambda p: p["price"])
    average_price = sum(p["price"] for p in products) / len(products)
    return {
        "cheapest_product": cheapest,
        "most_expensive_product": expensive,
        "average_price": average_price
    }
```
